```python
# Filter the DataFrame to include only the top 5 activities
top_activities = ['Swimming', 'Surfing', 'Fishing', 'Spearfishing', 'Wading']
filtered_df = df[df['activity'].isin(top_activities)]

# Group the filtered DataFrame by 'Activity' and 'Age' and calculate the count of incidents for each combination
incident_counts_by_activity_and_age = filtered_df.groupby(['activity', 'age']).size()

# Convert the Series to a DataFrame and reset the index
incident_counts_by_activity_and_age = incident_counts_by_activity_and_age.reset_index(name='count')

# Create a pivot table
pivot_table = incident_counts_by_activity_and_age.pivot_table(index='activity', columns='age', values='count', fill_value=0)

# Display the pivot table
print(pivot_table)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt=".0f", linewidths=.5)
plt.title('Number of Shark Attack Incidents by Age and Activity')
plt.xlabel('Age')
plt.ylabel('Activity')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
types = ['Provoked', 'Unprovoked']
sexes = ['Female', 'Male']
counts = {'Provoked': {'Female': 16, 'Male': 260},
          'Unprovoked': {'Female': 467, 'Male': 2575}}

# Grouped bar chart
fig, ax = plt.subplots()

# Plot bars
bar_width = 0.35
index = range(len(types))
opacity = 0.8

bars = []
for i, sex in enumerate(sexes):
    counts_sex = [counts[t][sex] for t in types]
    bars.append(ax.bar([p + i * bar_width for p in index], counts_sex, bar_width, alpha=opacity, label=sex))

# Add count labels on the bars
for bar in bars:
    for b in bar:
        plt.text(b.get_x() + b.get_width() / 2, b.get_height(), str(int(b.get_height())), ha='center', va='bottom')

# Set labels and title
ax.set_xlabel('Type')
ax.set_ylabel('Count')
ax.set_title('Shark Attack Incidents by Type and Sex')
ax.set_xticks([p + bar_width / 2 for p in index])
ax.set_xticklabels(types)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter the DataFrame to include only the top 5 activities
top_activities = ['Swimming', 'Surfing', 'Fishing', 'Spearfishing', 'Wading']
filtered_df = df[df['activity'].isin(top_activities)]

# Group the filtered DataFrame by 'Activity' and 'Sex' and calculate the count of incidents for each combination
incident_counts_by_activity_and_sex = filtered_df.groupby(['activity', 'sex']).size().unstack(fill_value=0)

# Plot the grouped bar chart
incident_counts_by_activity_and_sex.plot(kind='bar', stacked=False, figsize=(10, 6))

# Add labels and title
plt.title('Shark Attack Incidents by Top 5 Activities and Sex')
plt.xlabel('Activity')
plt.ylabel('Count')

# Show the plot
plt.xticks(rotation=45)
plt.legend(title='Sex')
plt.tight_layout()
plt.show()
```
